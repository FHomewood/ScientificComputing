```python
#Libraries
from scipy.integrate import odeint
from numpy import arange
from matplotlib.pyplot import plot, show, legend, figure, title, xlabel, ylabel
#

#Constants
a = 5.0
b = 0.9
c = [8.2, 8.3]

u0 = 0.0
v0 = 1.0
w0 = 2.0
#

#Subroutines
def L(y, t, a, b, c):
    u,v,w = y
    U = -a*(u-v)
    V = c*u-v-u*w
    W = -b*w + u*v
    return U,V,W
#

#Computation
for i in range(2):
    x = arange(0,10,0.01)
    y = odeint(L,[u0,v0,w0],x, args = (a,b,c[i]))
    figure(0)
    plot(x,y[:,0], ["k-", "r-"][i], label = ["c = 8.2","c = 8.3"][i])
    title("u(t) against t")
    xlabel("x -axis")
    ylabel("y -axis")
    legend()
    
    figure(1)
    plot(y[:,0],y[:,1], ["k-", "r-"][i], label = ["c = 8.2","c = 8.3"][i])
    title("u(t) against v(t)")
    xlabel("x -axis")
    ylabel("y -axis")
    legend()
show()
#
```
