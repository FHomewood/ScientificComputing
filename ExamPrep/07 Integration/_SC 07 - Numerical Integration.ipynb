```python
#Import necessary codes
import scipy as sci
from scipy import integrate
import numpy as np
import pylab as plt
```

# Question 1

```python
#Define the two integrals to be solved.
def C(x):
    return (np.cos(np.pi*x**2/2.))
def S(x):
    return (np.sin(np.pi*x**2/2.))
```

```python
#Set up the vector of values where we want the integrals evaluated and zero-filled vectors to later hold the integrals’ values

v=np.linspace(0,10,1000)

I=np.zeros(len(v))
ISa=np.zeros(len(v))
ICa=np.zeros(len(v))
```

```python
#Start a loop over the v values and calculate the integrals using the in-built function and store the values.
for n in range(0,len(v)-1):
    IC,errC=sci.integrate.quad(C,0.,v[n])
    IS,errS=sci.integrate.quad(S,0.,v[n])
    
    I[n]=0.5*((IC+0.5)**2+(IS+0.5)**2)
    ICa[n]=IC
    ISa[n]=IS

'''Note: doing this computationally efficiently would mean not re-doing
the integrals from the start, but re-using the part already calculated to
get the next point, etc. This could be achieved by a minor modification
of the above code, by replacing the low limits in quad by v[n − 1] and
adding the result to the previous integral to v[n − 1].'''
```

```python
plt.plot(v,I,'bo', label='I/I0')
plt.plot(v,ICa, label='C(v)')
plt.plot(v,ISa, label='S(v)')
plt.legend(loc='best')
plt.title('Plot of functions vs. v')
plt.xlabel('v')
plt.ylabel('I')
plt.show()
```
