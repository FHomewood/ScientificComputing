```python
from numpy import pi, arange, log, exp, sqrt, sin, cos, linspace
from matplotlib.pyplot import plot, show, xlim, ylim, xlabel, ylabel, title
from scipy.integrate import quad, odeint
from scipy.optimize import fsolve
```

```python
def g(c):
    def f(t):
        return 1-exp(-c*t) - t
    return f
x = fsolve(g(2),2)
print("The solution is {:.5f}".format(x[0]))

c = arange(0,3,0.01)
x =[]
for i in c:
    x.append(fsolve(g(i),1))

plot(c,x)
ylabel("x")
xlabel("c")
show()
```

```python
c = 0.005
g = 9.81
# q = x
# w = x'
# e = y
# r = y'

def L(y, t):
    q,w,e,r = y
    Q = w
    W = -c * w * sqrt(w**2 + r**2)
    E = r
    R = -c * r * sqrt(w**2 + r**2) - g
    return Q,W,E,R

#Computation
x = linspace(0,30,1000)
for alpha in [10,20,80]:
    y  = odeint(L,[0,1000*cos(alpha*pi/180),0,1000*sin(alpha*pi/180)],x)
    plot(y[:,0],y[:,2], "k-")
    title(str(alpha) + " degree trajectory")
    xlim(0,800)
    ylim(0,800)
    show()
```
