
## Question 1
### i)

$Let \ q = \theta_1 \\ Let \ w = \dot{\theta_1} \\ Let \ e = \theta_2 \\ Let \ r = \dot{\theta_2} \\ $

Using this new set of variables we can write the two second order differential equations as a set of four coupled first order differential equations

$ \dot{q} = \dot{\theta_1} = w \\  \dot{w} = \ddot{\theta_1} = \beta sin(\theta_2 - \theta_1) - b\dot{\theta_1} = \beta sin(e - q) - bw\\ 
  \dot{e} = \dot{\theta_2} = r \\  \dot{r} = \ddot{\theta_2} = \beta sin(\theta_1 - \theta_2) - b\dot{\theta_2} = \beta sin(q -e) - br\\ \\ $
  
  Simplifying these gives our set of four differential equations

$ \dot{q} = w \\  \dot{w} = \beta sin(e - q) - bw\\ 
  \dot{e} = r \\  \dot{r} = \beta sin(q -e) - br\\ $


### ii)

```python
b = 0.1
B = 5.0



def L(y, t):
    q,w,e,r = y
    Q = w
    W = B*sin(e-q) - b*w
    E = r
    R = B*sin(q-e) - b*r
    return Q,W,E,R

#Computation
x = linspace(0,30,1000)
y  = odeint(L,[1.,0.5,0.,0.],x)
for i in [0,1,2,3]:
    plot(x,y[:,i], "k-")
    title(["$Graph \ of \ \Theta_1 \ against \ time$",
           "$Graph \ of \ \dot{\Theta_1} \ against \ time$",
           "$Graph \ of \ \Theta_2 \ against \ time$",
           "$Graph \ of \ \dot{\Theta_2} \ against \ time$"][i])
    show()
```

```python
#Define Subroutines
def diff(f,x,h=0.00000000001):
    return (f(x+h)-f(x-h))/2/h
def cubicF(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d
#

#Define Raw data
x = [0.,1.525 ,3.050 ,4.575 ,6.10  ,7.625 ,9.150]
y = [1.,0.8617,0.7385,0.6292,0.5328,0.4481,0.3741]
#

x0 = np.arange(0,9.150,0.01)
interpolate  = interp1d(x,y, kind = 'cubic')
Coe, Cov = curve_fit(cubicF,x,y)

plot(x,y,'kx',label ="Raw Data")
plot(x0,cubicF(x0,Coe[0],Coe[1],Coe[2],Coe[3]),'b-',label ="Curve Fitting")
plot(x0,interpolate(x0),'r-',label ="Interpolation")
legend()
show()

for h in [2,4,8]:
    print("when h = " + str(h) + "; rho = {:.5f}".format(cubicF(h,Coe[0],Coe[1],Coe[2],Coe[3])))

print("")
print ("using the cubic spline gives a value of {:.5f}"
       .format(float(interpolate(4))))

print("")
print("the relative error is {:.3f}% for the cubic fit (polynomial fit of maximum degree)".format(abs(100*0.67/interpolate(4)-100)))
print("and {:.3f}% for the cubic splines".format(abs(100*0.67/cubicF(4,Coe[0],Coe[1],Coe[2],Coe[3])-100)))
print("")
print("the absolute error is {:.3e} for the cubic fit (polynomial fit of maximum degree)".format(abs(interpolate(4)-0.67)))
print("and {:.3e} for the cubic splines".format(abs(cubicF(4,Coe[0],Coe[1],Coe[2],Coe[3])-0.67)))
```
